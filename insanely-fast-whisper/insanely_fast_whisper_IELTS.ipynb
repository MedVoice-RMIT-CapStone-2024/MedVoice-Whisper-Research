```python
import subprocess

def run_cli():
    # Define the command as a string
    command = "insanely-fast-whisper --file-name https://storage.googleapis.com/medvoice-sgp-audio-bucket/Cambridge-IELTS-16-Listening-Test-1-Part-1.mp3 --model-name openai/whisper-large-v2 --language en --hf-token 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 --num-speakers 3"

    # Use subprocess to run the command
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    # Get the output and error messages, if any
    stdout, stderr = process.communicate()

    # Decode the output and error messages from bytes to string
    stdout = stdout.decode()
    stderr = stderr.decode()

    # Print the output and error messages
    print("Output:", stdout)
    if stderr:
        print("Error:", stderr)

# Call the function to run the CLI
run_cli()
```

```python
import json

def format_json(file_path):
    with open(file_path, 'r') as json_file:
        json_data = json.load(json_file)
    formatted_json = json.dumps(json_data, indent=4)
    print(formatted_json)

    with open('formatted_json', 'w') as output_file:
      output_file.write(formatted_json)

format_json("output.json")
```

```python
import json

def who_speak_when(file_path):
    with open(file_path, 'r') as json_file:
      data = json.load(json_file)

    # Initialize a dictionary to store the combined text for each speaker
    combined_text = {}

    # Iterate over each speaker in the data
    for speaker in data["speakers"]:
        # If the speaker is already in the dictionary, append their new text
        if speaker["speaker"] in combined_text:
            combined_text[speaker["speaker"]] += speaker["text"]
        # If the speaker is not in the dictionary, add them with their text
        else:
            combined_text[speaker["speaker"]] = speaker["text"]

    # Print the combined text for each speaker
    for speaker, text in combined_text.items():
        print(f"{speaker}: {text}")

who_speak_when("output.json")
```
